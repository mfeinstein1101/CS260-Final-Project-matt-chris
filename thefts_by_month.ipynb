```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
data = np.zeros((2, 61))

with open('data/Thefts_By_Month.csv') as file:
    reader = csv.reader(file)
    for line in reader:
        print(line)
        if line[0] == 'idx':
            data[0,:] = line[1:]
        if line[0] == 'United States':
            for i in range(1, 62):
                if line[i] != '':
                    data[1, i-1] = line[i]

plt.plot(data[0,:49], data[1,:49])
```
